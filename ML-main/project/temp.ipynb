```python
import pandas as pd
from string import ascii_uppercase as alphabet
import pickle
```

```python
all_tables =pd.read_html('https://en.wikipedia.org/wiki/UEFA_Euro_2024')
```

```python
for i in range (18,60,7):
    print(all_tables[i])
```

```python

dict_tables = {}
for letter,i in zip(alphabet,range (18,60,7)):
    df = all_tables[i]
    df.rename(columns={'Teamvte':'Team'},inplace=True)
    df.pop('Qualification')
    print(letter,i)
    dict_tables[f'Group {letter}'] = df
```

```python
with open('dict_tables','wb') as output :
    pickle.dump(dict_tables,output)
```
